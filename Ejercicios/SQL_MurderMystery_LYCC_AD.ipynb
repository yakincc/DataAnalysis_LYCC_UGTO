### Pista 1
Sabemos que hubieron 2 testigos:
1. S/N, vive en "Northwestern Dr"
2. Annabel, vive en "Franklin Ave"

Podemos buscar la información de los dos posibles testigos en la tabla 'person'

**Testigo 1**
Vive en la última casa de Northwestern Dr

```python
query = """
SELECT *
FROM person
WHERE address_street_name = 'Northwestern Dr'
ORDER BY address_number DESC
LIMIT 1
"""

first_withness_info = pd.read_sql(query, conn)
first_withness_info
```

**Testigo 2 : Annabel**

```python
query = """
SELECT *
FROM person
WHERE address_street_name = 'Franklin Ave' AND name LIKE 'Annabel%'
"""

second_withness_info = pd.read_sql(query, conn)
second_withness_info
```
